```python
import pandas as pd
```

```python
df = pd.read_csv("data/silver_20250322_Airline_Reviews_SentimentImageLabelsMerged.csv")
df
```

#### Correct Overall_Rating 10 bug

```python
df_20250324 = pd.read_csv("data/bronze_20250324_Airline_Reviews_withImageUrls.csv")
df_20250324
```

```python
df_20250324.rename(columns={"Unnamed: 0": "RowId"}, inplace=True)
df_20250324
```

```python
right_columns_needed = ['RowId','Overall_Rating']

df_merged= pd.merge(df, df_20250324[right_columns_needed], on='RowId', how='inner', suffixes=('_left', '_right'))
df_merged
```

```python
df_merged.rename(columns={"Overall_Rating_right": "Overall_Rating"}, inplace=True)
df_merged.drop(columns=['Overall_Rating_left'], inplace=True)
df_corrected_overallrating10 = df_merged
df_corrected_overallrating10
```
