```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/silver_20250322_Airline_Reviews_LostLuggageLabel.csv")
print(len(df))
df.head()
```

```python
df['Review Date'] = pd.to_datetime(df['Review Date'], format="mixed")
print(len(df))
df.head()
```

#### Correct Overall_Rating 10 bug

```python
df_20250324 = pd.read_csv("../data/bronze_20250324_Airline_Reviews_withImageUrls.csv")
print(len(df_20250324))
df_20250324.head()
```

```python
df_20250324.rename(columns={"Unnamed: 0": "RowId"}, inplace=True)
print(len(df_20250324))
df_20250324.head()
```

```python
df_20250324['Review Date'] = pd.to_datetime(df_20250324['Review Date'], format="mixed")
print(len(df_20250324))
df_20250324.head()
```

```python
df_20250324_to_correct = df_20250324[df_20250324['Overall_Rating'] == '10']
print(len(df_20250324_to_correct))
df_20250324_to_correct.head()
```

```python
right_columns_needed = ['RowId','Airline Name','Review_Title','Review Date','Overall_Rating']

df_merged= pd.merge(df, df_20250324_to_correct[right_columns_needed], on=['Airline Name','Review_Title','Review Date'], \
                    how='left', suffixes=('', '_right'))

print(len(df_merged))
df_merged[['RowId','Airline Name','Review_Title','Review Date','Overall_Rating','Overall_Rating_right','RowId_right']].head()
```

```python
df_merged.drop(columns=['RowId_right'], inplace=True)

#df_merged.rename(columns={"Overall_Rating_right": "Overall_Rating"}, inplace=True)
df_merged['Overall_Rating'] = np.where(df_merged['Overall_Rating_right'].isna(), df_merged['Overall_Rating'], df_merged['Overall_Rating_right'])

df_merged.drop(columns=['Overall_Rating_right'], inplace=True)

df_corrected_overallrating10 = df_merged

print(len(df_corrected_overallrating10))
df_corrected_overallrating10[['RowId','Airline Name','Review_Title','Review Date','Overall_Rating']].head()
```

#### Check corrected

```python
df_corrected_overallrating10[df_corrected_overallrating10['Overall_Rating'] == '10']

print(len(df_corrected_overallrating10[df_corrected_overallrating10['Overall_Rating'] == '10']))
df_corrected_overallrating10[df_corrected_overallrating10['Overall_Rating'] == '10'].head()
```

#### Persist to CSV file
