```python
import pandas as pd
```

```python
df_20250322 = pd.read_csv('data/bronze_20250322_Airline_Reviews_withImageUrls.csv')
df_20250322
```

```python
df_20250322['Review Date'] = df_20250322['Review Date'].str.replace('th','')
df_20250322['Review Date'] = df_20250322['Review Date'].str.replace('st','')
df_20250322['Review Date'] = df_20250322['Review Date'].str.replace('rd','')
df_20250322['Review Date'] = df_20250322['Review Date'].str.replace('Augu','August')
```

```python
df_20250322['Review Date']
```

```python
df_20250322['Review Date'] = pd.to_datetime(df_20250322['Review Date'], format='mixed')
```

```python
df_20250322
```

```python
df_20250324 = pd.read_csv('data/bronze_20250324_Airline_Reviews_withImageUrls.csv')
df_20250324
```

```python
df_20250324['Review Date'] = df_20250324['Review Date'].str.replace('th','')
df_20250324['Review Date'] = df_20250324['Review Date'].str.replace('st','')
df_20250324['Review Date'] = df_20250324['Review Date'].str.replace('rd','')
df_20250324['Review Date'] = df_20250324['Review Date'].str.replace('Augu','August')
```

```python
df_20250324['Review Date']
```

```python
df_20250324['Review Date'] = pd.to_datetime(df_20250324['Review Date'], format='mixed')
```

```python
df_20250324
```

```python
df_20250322_grouped = df_20250322.groupby(['Airline Name']).agg(review_count=('Airline Name', 'count'), max_review_date=('Review Date', 'max'))\
    .sort_values(by='review_count', ascending=False)
df_20250322_grouped
```

```python
df_20250324_grouped = df_20250324.groupby(['Airline Name']).agg(review_count=('Airline Name', 'count'), max_review_date=('Review Date', 'max'))\
    .sort_values(by='review_count', ascending=False)
df_20250324_grouped
```

```python
df_20250324_22_merged = df_20250324_grouped.merge(df_20250322_grouped, how='left', on=['Airline Name'], suffixes=('_0324', '_0322'))
```
